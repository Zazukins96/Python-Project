```python
df=pd.read_excel('pyproject.xlsx')
df
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
df
```

```python
# 1.How many are there in each Team and the percentage splitting with respect to the total employees.
total_employees=df['Name'].count()
team_count=df['Team'].value_counts()

percentage_split=(team_count/total_employees)*100

team_details=pd.DataFrame({'TEAM':team_count.index,'NUMBER OF EMPLOYEES':team_count.values,'PERCENTAGE_SPLIT':percentage_split.values})
team_details
```

```python
### 2.segragate the employees w.r.t different positions.
# number of employees in each position
position_count=df['Position'].value_counts()
position_details=pd.DataFrame({'Position':position_count.index,'Number of employees':position_count})
print('Position Details')
print(position_details)
```

```python
### 3.find from which age group most of the employees belongs to
age_data=df['Age'].value_counts()
age_details=pd.DataFrame({'Age':age_data.index,'Number of employees':age_data.values})
print('Age Details')
print(age_details)

# Histogram representation
plt.hist(df['Age'],bins=[0,10,20,30,40,50],edgecolor='orange')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
### 4.find out under which team and position,spending in terms of salay is high.
max_salary_spending=df.groupby(['Team','Position'])['Salary'].sum().idxmax()
print(max_salary_spending)
```

```python
### 5.find if there is any correlation between age and salary,represent it visually
# scatterplot
sns.scatterplot(x='Age',y='Salary',hue='Position',data=df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age and salary correlation')
plt.show()
# heatmap
new=df[['Age','Salary']]
corr=new.corr()
sns.heatmap(corr,annot=True,cmap='cool')
plt.title('Correlaton Matrix')
plt.show()
print('The the correlation coefficient is:',df['Age'].corr(df['Salary']))
```
